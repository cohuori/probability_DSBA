```python
import pandas as pd

problem_texts = [
    # CATEGORY 1
    "Class of 100 students: Men/Women and Taking/Not taking Economics table. Find probabilities: woman, taking economics, man or taking economics, woman and taking economics.",
    "Families with three children, probability of boy is 52%. Use outcome table BBB, BBG, ..., GGG. Find probabilities: exactly 2 girls, at least 2 girls, at least one child of each sex, middle child opposite to the other two.",
    "Take 3 cards from a 36-card deck with replacement. Describe elementary outcomes and probabilities. Find P(queen, king, ace) and P(king, king, ace).",
    "Take 3 cards from a 36-card deck without replacement. Answer the same questions as in the previous problem.",
    "Find unions and intersections of given events A, B, C. Decide when one event is a subset of another.",
    "Basic operations with events: mutually exclusive events, union formula, complement of intersection.",
    "Survey of houses: holes in roof, broken windows, both problems. Find probabilities of union, roof holes only, exactly one problem, no problems.",
    "Pick a random word from the sentence: 'Suppose a word is picked at random from this sentence.' Find sample space and probabilities involving letters and vowels.",
    "In a group of students, 25% smoke hookah, 60% drink alcohol, 15% do both. Find fraction with at least one bad habit.",
    "20 families near Wolfenstein Castle: 4 have 1 child, 8 have 2, 5 have 3, 3 have 4. If we meet a random local child, find p1, p2, p3, p4.",
    "Roll a red die and a green die. Find the probability that the red die is larger than the green die.",
    "Two dice are rolled. Find probabilities: the numbers differ by 1 or less; the maximum is 5 or larger.",
    "Galileo dice problem: three dice. Compare probability of total 9 and total 10.",

    # CATEGORY 2
    "n friends sit around a table in random order. Find the probability that A sits next to B.",
    "Box with 10 white balls and 6 black balls. Choose 4 without replacement. Find probability of at least one black ball and exactly two black balls.",
    "Closet contains n pairs of boots. Randomly select 2r boots. Find probability of no complete pairs and exactly one complete pair.",
    "Birthday problem: among 50 students, find probability that at least two have the same date of birth.",
    "Box with 28 black balls and 4 white balls. Choose 10. Find probability of at least one white, exactly one white, at least two white, exactly two white.",
    "52-card deck, 6 cards drawn. Find probability that the king of spades appears and probability that all four suits appear.",

    # CATEGORY 3
    "Two M-sided dice are thrown. Find probability that the sum is equal to i.",
    "n distinguishable balls are randomly placed into m boxes. Find probability that all boxes are non-empty.",
    "A shelf has 12 books in a row. Choose 5 books randomly. Find probability that no pair of adjacent books is chosen.",
    "Lucky tram ticket: six-digit number is special if sum of first three digits equals sum of last three digits. Find probability of getting a lucky ticket."
]

ranks = (
    [1] * 13 +
    [2] * 6 +
    [3] * 4
)

topics = (
    ["Category 1"] * 13 +
    ["Category 2"] * 6 +
    ["Category 3"] * 4
)

problems = pd.DataFrame({
    "problem_id": list(range(1, len(problem_texts) + 1)),
    "topic": topics,
    "problem_text": problem_texts,
    "rank": ranks
})

display(problems)
```

```python
students = pd.read_csv("students.csv")
#display(students)
```

```python
def assign_problems(
    students_df,
    problems_df,
    n_students,
    min_rank=None,
    max_rank=None,
    seed=None
):
    if seed is not None:
        random.seed(seed)

    # Filter problems by rank
    available_problems = problems_df.copy()

    if min_rank is not None:
        available_problems = available_problems[available_problems["rank"] >= min_rank]

    if max_rank is not None:
        available_problems = available_problems[available_problems["rank"] <= max_rank]

    # Check enough students
    if n_students > len(students_df):
        raise ValueError("You asked for more students than you have in the list.")

    # Check enough problems
    if n_students > len(available_problems):
        raise ValueError("Not enough problems in the selected rank range.")

    # Randomly choose students and problems
    chosen_students = students_df.sample(n=n_students, random_state=seed).reset_index(drop=True)
    chosen_problems = available_problems.sample(n=n_students, random_state=seed).reset_index(drop=True)

    # Combine
    result = pd.concat([chosen_students, chosen_problems], axis=1)

    return result
```

```python
assignment = assign_problems(
    students_df=students,
    problems_df=problems,
    n_students=3,
    min_rank=1,
    max_rank=1,
    seed=None
)

display(assignment)
```
