# Motivation

Probability starts with counting.

If all outcomes are equally likely,

$$
P(A)=\frac{|A|}{|\Omega|}.
$$

Therefore, many probability problems reduce to counting.

## Example

A password consists of 4 digits.

How many passwords exist?

We have

$$
10^4=10000.
$$

The probability of guessing correctly in one attempt is

$$
\frac{1}{10000}.
$$
