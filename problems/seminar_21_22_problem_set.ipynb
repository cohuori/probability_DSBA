## Theory Box: Confidence Intervals for Mean and Variance of a Normal Distribution

**Model.** Let \(X_1,X_2,\ldots,X_n\) be a random sample from the normal distribution

\[
X_i\sim \mathcal N(\mu,\sigma^2),
\qquad i=1,\ldots,n.
\]

### Theorem 1: Confidence Interval for the Mean when Variance is Known

If \(\sigma^2\) is known, then

\[
\frac{\overline X-\mu}{\sigma/\sqrt n}\sim \mathcal N(0,1).
\]

Hence, a \((1-\alpha)\times 100\%\) confidence interval for \(\mu\) is

\[
\overline X
\pm
z_{\alpha/2}\frac{\sigma}{\sqrt n},
\]

where \(z_{\alpha/2}\) is the upper \(\alpha/2\) quantile of the standard normal distribution.

### Theorem 2: Confidence Interval for the Mean when Variance is Unknown

If \(\sigma^2\) is unknown, then

\[
\frac{\overline X-\mu}{S/\sqrt n}\sim t_{n-1}.
\]

Therefore, a \((1-\alpha)\times 100\%\) confidence interval for \(\mu\) is

\[
\overline X
\pm
t_{\alpha/2,n-1}\frac{S}{\sqrt n},
\]

where \(t_{\alpha/2,n-1}\) is the upper \(\alpha/2\) quantile of the \(t\)-distribution with \(n-1\) degrees of freedom.

### Theorem 3: Confidence Interval for the Variance

Under normality,

\[
\frac{(n-1)S^2}{\sigma^2}\sim \chi^2_{n-1}.
\]

Hence, a \((1-\alpha)\times 100\%\) confidence interval for \(\sigma^2\) is

\[
\left(
\frac{(n-1)S^2}{\chi^2_{\alpha/2,n-1}},
\;
\frac{(n-1)S^2}{\chi^2_{1-\alpha/2,n-1}}
\right),
\]

where \(\chi^2_{p,n-1}\) is the \(p\)-quantile of the chi-squared distribution with \(n-1\) degrees of freedom.

**Remark.** Confidence intervals for \(\sigma\) can be obtained by taking square roots of the limits of the interval for \(\sigma^2\).
:::

A chiropterologist believes that the Kitti's hog-nosed bats in a new geographical region under study have a different average weight than \(1.9\) grams. A sample of \(10\) bats weighed in grams in the new region is shown below:

\begin{array}{c|cccccccccc}
\text{Weight (g)}
& 1.90 & 2.24 & 2.13 & 2.00 & 1.54 & 1.96 & 1.79 & 2.18 & 1.81 & 2.30
\end{array}
\]

Use the **confidence interval method** to test the claim that the **mean weight** for all bumblebee bats is not \(1.9\) g using a **10\% level of significance**.

## Theory Box: Confidence Intervals for the Difference of Means

**Model.** Suppose we have two independent samples:

\[
X_1,\ldots,X_{n_1}\sim \mathcal N(\mu_1,\sigma_1^2),
\qquad
Y_1,\ldots,Y_{n_2}\sim \mathcal N(\mu_2,\sigma_2^2).
\]

### Theorem 4: Difference of Means, Equal Variances

If \(\sigma_1^2=\sigma_2^2=\sigma^2\), define the pooled variance

\[
S_p^2
=
\frac{(n_1-1)S_X^2+(n_2-1)S_Y^2}{n_1+n_2-2}.
\]

Then

\[
\frac{
(\overline X-\overline Y)-(\mu_1-\mu_2)
}{
S_p\sqrt{\frac1{n_1}+\frac1{n_2}}
}
\sim t_{n_1+n_2-2}.
\]

Hence, a \((1-\alpha)\times 100\%\) confidence interval for \(\mu_1-\mu_2\) is

\[
\overline X-\overline Y
\pm
t_{\alpha/2,n_1+n_2-2}
S_p\sqrt{\frac1{n_1}+\frac1{n_2}}.
\]

### Theorem 5: Difference of Means, Unequal Variances — Welch's Approximation

If variances are not assumed equal, then

\[
\frac{
(\overline X-\overline Y)-(\mu_1-\mu_2)
}{
\sqrt{\frac{S_X^2}{n_1}+\frac{S_Y^2}{n_2}}
}
\sim t_{\nu},
\]

where the approximate degrees of freedom \(\nu\) are given by

\[
\nu
=
\frac{
\left(
\dfrac{S_X^2}{n_1}
+
\dfrac{S_Y^2}{n_2}
\right)^2
}{
\dfrac{
\left(\dfrac{S_X^2}{n_1}\right)^2
}{n_1-1}
+
\dfrac{
\left(\dfrac{S_Y^2}{n_2}\right)^2
}{n_2-1}
}.
\]

Thus, a \((1-\alpha)\times 100\%\) confidence interval for \(\mu_1-\mu_2\) is

\[
\overline X-\overline Y
\pm
t_{\alpha/2,\nu}
\sqrt{
\frac{S_X^2}{n_1}
+
\frac{S_Y^2}{n_2}
}.
\]
:::
