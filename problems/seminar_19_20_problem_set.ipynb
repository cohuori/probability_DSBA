## Theory Box: Least Squares Estimator (LSE)

**Definition.** Consider the linear regression model

$$
Y = X\beta + \varepsilon,
$$

where:

- $Y$ is an $n\times 1$ vector of observations;
- $X$ is an $n\times p$ matrix of known predictors, called the design matrix;
- $\beta$ is a $p\times 1$ vector of unknown parameters;
- $\varepsilon$ is an $n\times 1$ vector of random errors, typically assumed to satisfy

$$
\mathbb E[\varepsilon]=0,
\qquad
\operatorname{Var}(\varepsilon)=\sigma^2 I_n.
$$

The **least squares estimator** of $\beta$ minimizes the sum of squared residuals

$$
S(\beta)=(Y-X\beta)^\top(Y-X\beta).
$$

The solution is

$$
\widehat\beta=(X^\top X)^{-1}X^\top Y,
$$

provided that $X^\top X$ is invertible.

**Properties of the least squares estimator:**

1. **Unbiasedness:**

$$
\mathbb E[\widehat\beta]=\beta.
$$

2. **Variance-covariance matrix:**

$$
\operatorname{Var}(\widehat\beta)=\sigma^2(X^\top X)^{-1}.
$$

3. **Efficiency (Gauss--Markov theorem):** among all linear unbiased estimators, $\widehat\beta$ has the minimum variance.

4. **Normality:** if the errors are normal, then

$$
\widehat\beta\sim \mathcal N\!\left(\beta,\sigma^2(X^\top X)^{-1}\right).
$$
:::

## Problem 1

Suppose that you are given independent observations $y_1$, $y_2$ and $y_3$ such that

$$
\begin{aligned}
y_1 &= 3\alpha+\beta+\varepsilon_1,\\
y_2 &= \alpha-2\beta+\varepsilon_2,\\
y_3 &= -\alpha+2\beta+\varepsilon_3.
\end{aligned}
$$

The random variables $\varepsilon_i$, for $i=1,2,3$, are normally distributed with mean $0$ and variance $4$.

1. Find the least squares estimators of the parameters $\alpha$ and $\beta$, and verify that they are unbiased estimators.
2. Calculate the variance of the estimator of $\alpha$.
:::

## Problem 2

Suppose that you are given observations $y_1$, $y_2$ and $y_3$ such that

$$
\begin{aligned}
y_1 &= 2\alpha+\beta+\varepsilon_1,\\
y_2 &= -\alpha-\beta+\varepsilon_2,\\
y_3 &= -\alpha+\beta+\varepsilon_3.
\end{aligned}
$$

The random variables $\varepsilon_i$, for $i=1,2,3$, are independent and normally distributed with mean $0$ and variance $\sigma^2$, while $\alpha$ and $\beta$ are unknown parameters.

Find the least squares estimator of $\beta$, verify that it is an unbiased estimator of $\beta$, and find the variance of this estimator.
:::

## Problem 4

Let $X_1,\ldots,X_n$ be a sample from a uniform distribution on the interval $[0,\theta]$.

Compare the following estimators of the parameter $\theta$ in the mean square error sense, that is, decide which one is the most efficient:

$$
2\overline X,
\qquad
(n+1)X_{(1)},
\qquad
\frac{n+1}{n}X_{(n)}.
$$

## Problem 6

In the quadrilateral $ABCD$, independent measurements of the angles $ABD$, $DBC$, $ABC$, $BCD$, $CDB$, $BDA$, $CDA$, $DAB$ (in degrees) gave the results

$$
50.78,
\quad
30.25,
\quad
78.29,
\quad
99.57,
\quad
50.42,
\quad
40.59,
\quad
88.87,
\quad
89.86,
$$

respectively.

Assuming that the measurement errors are normally distributed according to $\mathcal N(0,\sigma^2)$, find the optimal estimates of the angles

$$
\beta_1=ABD,
\qquad
\beta_2=DBC,
\qquad
\beta_3=CDB,
\qquad
\beta_4=BDA,
$$

and the unknown variance $\sigma^2$.
