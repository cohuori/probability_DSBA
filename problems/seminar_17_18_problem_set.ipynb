## Theory: Maximum Likelihood Estimator (MLE)

**Definition.** Let $X_1, \dots, X_n$ be a random sample from a probability distribution with density or mass function $p_\theta(x)$ depending on a parameter $\theta$. The likelihood function is defined as

$$
L(\theta)=\prod_{i=1}^{n}p_\theta(X_i).
$$

The log-likelihood function is

$$
\ell(\theta)=\log L(\theta)=\sum_{i=1}^{n}\log p_\theta(X_i).
$$

A maximum likelihood estimator of $\theta$ is any value $\widehat\theta$ that maximizes $L(\theta)$, or equivalently $\ell(\theta)$:

$$
\widehat\theta=\arg\max_\theta L(\theta).
$$

**Properties of the MLE.** Under suitable regularity conditions, the MLE has the following properties:

- **Consistency:** $\widehat\theta\to\theta$ in probability as $n\to\infty$.
- **Asymptotic efficiency:** the MLE achieves the Cramér--Rao lower bound asymptotically.
- **Invariance:** if $g$ is a differentiable function, then $g(\widehat\theta)$ is the MLE of $g(\theta)$.
:::

::: {.callout-tip collapse="true"}
## Theory: Method of Moments Estimator (MME)

**Definition.** Let $X_1,\dots,X_n$ be a random sample from a distribution with parameter $\theta$. The theoretical moments of order $k$ are

$$
\mu_k'(\theta)=\mathbb E[X^k].
$$

The method of moments equates the first $r$ sample moments to the corresponding theoretical moments and solves for the parameters. For parameters $\theta=(\theta_1,\dots,\theta_r)$, solve

$$
\frac1n\sum_{i=1}^{n}X_i^k=\mu_k'(\theta),\qquad k=1,\dots,r.
$$

The solution $\widehat\theta$ is called the method of moments estimator.

**Properties of the MME.** Under suitable conditions, the MME has the following properties:

- **Consistency:** the MME is consistent under mild conditions, especially if the moments uniquely determine the distribution.
- **Simplicity:** often simpler to compute than the MLE, since it avoids maximizing complex likelihood functions.
- **Efficiency:** MMEs are generally less efficient than MLEs, meaning they may have larger asymptotic variance.
:::

## Category 1 — Problem 3

A random sample $\{X_1,X_2,\ldots,X_n\}$ is drawn from the following probability distribution:

$$
f(x;\theta)=
\begin{cases}
\dfrac{2x}{1-\theta^2},& \theta\le x\le 1,\\[1ex]
0,&\text{otherwise.}
\end{cases}
$$

1. Derive the maximum likelihood estimator of $\theta$ and explain why this is a maximum.
2. If $n=4$, with $x_1=0.70$, $x_2=0.92$, $x_3=0.21$, and $x_4=0.34$, determine the maximum likelihood estimate of $\theta$.
3. As $n\to\infty$, would you expect the maximum likelihood estimate of $\theta$ to get closer to $\theta$? Briefly explain your answer.
