# Problem Set 13–14

*During the seminars, we will work on problems across three different categories of difficulty, with Category 1 being the easiest and Category 3 the most challenging.*


::: {.callout-tip collapse="true"}
## Weak Law of Large Numbers (WLLN)

Let $X_1, X_2, \dots, X_n$ be independent and identically distributed random variables with finite mean $\mu$. Then

$$
\overline{X}_n 
= \frac{1}{n} \sum_{i=1}^{n} X_i 
\xrightarrow{P} 
\mu
$$

as $n \to \infty$.
:::


::: {.callout-tip collapse="true"}
## Central Limit Theorem (CLT)

Let $X_1, X_2, \dots, X_n$ be independent and identically distributed random variables with finite mean $\mu$ and finite variance $\sigma^2 > 0$. Then

$$
\frac{\sum_{i=1}^{n} X_i - n\mu}{\sigma \sqrt{n}} 
\xrightarrow{d} 
\mathcal{N}(0, 1)
$$

as $n \to \infty$.
:::


::: {.callout-tip collapse="true"}
## Poisson Limit Theorem

Let $X_n \sim \operatorname{Binomial}(n, p_n)$, where $p_n \to 0$ and $n p_n \to \lambda > 0$ as $n \to \infty$. Then

$$
\lim_{n \to \infty} 
\Pr\bigl( X_n = k \bigr)
= \frac{e^{-\lambda} \lambda^k}{k!}, 
\quad k = 0, 1, 2, \dots
$$

that is, $X_n$ converges in distribution to a Poisson random variable with parameter $\lambda$.
:::


## Category 1

::: {.callout-tip collapse="true"}
## Category 1 — Problem 1

1800 dice are thrown. Find an approximate value for the probability that the total number of occurrences of 2 and 6 is not less than 620.
:::


::: {.callout-tip collapse="true"}
## Category 1 — Problem 2

When typing, the stenographer makes a mistake in the character with probability of $0.0005$. Find an approximate value for the probability that, when typing $10{,}000$ characters, the stenographer will make a mistake no more than three times.
:::


::: {.callout-tip collapse="true"}
## Category 1 — Problem 3

A bank teller serves customers standing in the queue one by one. Suppose that the service time $\xi_i$ for customer $i$ has mean ${\sf E}\xi_i=2$ minutes and ${\sf Var}(\xi_i)=1$. We assume that service times for different bank customers are independent. Let $\eta$ be the total time the bank teller spends serving 50 customers. Find

$$
{\sf P}(90<\eta<110).
$$
:::


::: {.callout-tip collapse="true"}
## Category 1 — Problem 4

The probability that a light bulb lasts more than 1000 hours is $\frac{1}{3}$. Estimate the probability that out of 1800 light bulbs, the lifetime of at least 580 bulbs exceeds 1000 hours.
:::


## Category 2

::: {.callout-note collapse="true"}
## Category 2 — Problem 1

A die is thrown $n$ times. Let the random variable $S_n$ correspond to the sum of the points obtained. Find $n$ such that

$$
{\sf P}\left(\left| \frac{S_n}{n}-3.5 \right|\geq 0.1\right) \geq 0.05.
$$

a) Using Chebyshev's inequality.

b) Using the Central Limit Theorem.
:::


::: {.callout-note collapse="true"}
## Category 2 — Problem 2

How many times should we throw a die if we want that the sum of points obtained was at least 4500 with probability $p\geq 0.975$? Use the Central Limit Theorem.
:::


## Category 3

::: {.callout-caution collapse="true"}
## Category 3 — Problem 1

Prove that

$$
\lim\limits_{n\to\infty} 
\frac{1}{e^n} 
\sum\limits_{k=0}^{n} 
\frac{n^k}{k!} 
= \frac{1}{2},
$$

using the Central Limit Theorem.
:::


::: {.callout-caution collapse="true"}
## Category 3 — Problem 2

Let $\{\xi_n\}$ be a sequence of independent identically distributed random variables with distribution $\mathcal{U}(0,1)$. Set

$$
M_n=\max \{\xi_1,\ldots,\xi_n \}.
$$

Prove that

$$
n(1-M_n) \xrightarrow[]{\sf d} {\sf Exp}(1).
$$
:::


::: {.callout-caution collapse="true"}
## Category 3 — Problem 3

Let $\{\xi_n,\ n\in\mathbb N\}$ be independent random variables with finite variance. Prove that for any $x\in\mathbb R$ the following limit exists:

$$
\lim_{n \to \infty} P(\xi_1+\cdots+\xi_n\le x)
$$

and that it equals $0$, $1$, or $\frac12$. Specify the conditions under which each of these values occurs.
:::
