# Problem Set 15–16: Estimators, Bias, Consistency and MSE

*During the seminars, we will work on problems across three different categories of difficulty, with Category 1 being the easiest and Category 3 the most challenging.*


## Category 1

::: {.callout-tip collapse="true"}
### Theory Review

**Unbiased Estimator.** An estimator $\hat{\theta}$ of a parameter $\theta$ is said to be **unbiased** if

$$
\mathbb{E}[\hat{\theta}] = \theta.
$$

**Consistent Estimator.** An estimator $\hat{\theta}_n$ of a parameter $\theta$ is said to be **consistent** if

$$
\hat{\theta}_n \xrightarrow{P} \theta \quad \text{as } n \to \infty,
$$

i.e., $\hat{\theta}_n$ converges in probability to $\theta$ as the sample size $n$ increases.

**Mean Squared Error (MSE).** The **mean squared error** of an estimator $\hat{\theta}$ of a parameter $\theta$ is defined as

$$
\mathrm{MSE}(\hat{\theta}) = \mathbb{E}\bigl[(\hat{\theta} - \theta)^2\bigr].
$$

It can be decomposed as

$$
\mathrm{MSE}(\hat{\theta})
=
\operatorname{Var}(\hat{\theta})
+
\left(\operatorname{Bias}(\hat{\theta})\right)^2.
$$

**Efficient Estimator.** An estimator $\hat{\theta}$ is called **efficient** if it is unbiased and achieves the lowest possible variance among all unbiased estimators of $\theta$. For regular problems, this variance lower bound is given by the Cramér--Rao lower bound.

**Sample Mean.** Given a sample $X_1,X_2,\ldots,X_n$, the **sample mean** $\bar X$ is defined as

$$
\bar X=\frac1n\sum_{i=1}^n X_i.
$$

**Sample Variance.** Given a sample $X_1,X_2,\ldots,X_n$, the **sample variance** $S^2$ is defined as

$$
S^2=\frac1{n-1}\sum_{i=1}^n (X_i-\bar X)^2.
$$

**Sample Proportion.** Given a sample of $n$ Bernoulli trials with success indicator variables $X_1,X_2,\ldots,X_n$, the **sample proportion** $\hat p$ of successes is defined as

$$
\hat p=\frac1n\sum_{i=1}^n X_i.
$$
:::

### Problem 1

Random variable $X$ assumes values $0$ and $1$, each with probability $1/2$. You have $9$ independent observations of $X$. Consider the following estimators of the population mean $\mu$:

- $\hat{\mu}_1=0.45$
- $\hat{\mu}_2=X_1$
- $\hat{\mu}_3=\overline{X}$
- $\hat{\mu}_4=\dfrac{2}{3}X_1+\dfrac{2}{3}X_2-\dfrac{1}{3}X_3$

Are these estimators biased? Which one is the most efficient?
:::

### Problem 3

Let $X_1,\ldots,X_n$ be a sample from some distribution with parameter $\sigma^2$. Assume additionally that

$$
\operatorname{Var}(X_1)=\sigma^2.
$$

Define the uncorrected sample variance as the statistic

$$
s^2=\frac1n\sum_{i=1}^n (X_i-\overline X)^2.
$$

Prove that:

1. $s^2=\overline{X^2}-\overline X^2$;
2. Is $s^2$ an unbiased estimator for $\sigma^2$?
:::

### Problem 6
