### Useful facts: Poisson distribution

Poisson distribution: $X\sim \operatorname{Poisson}(\lambda)$, where $X$ shows the number of occurrences in a Poisson process with an average number of events $\lambda$.

$$
P(X=k)=\frac{e^{-\lambda}\lambda^k}{k!},
\qquad k\in\mathbb N_0,
\qquad
\mathbb E(X)=\lambda,
\qquad
\operatorname{Var}(X)=\lambda.
$$

If $r$ is an average rate of occurrences, and $t$ is a time period, then

$$
P(X=k)=\frac{e^{-rt}(rt)^k}{k!},
\qquad k\in\mathbb N_0,
\qquad
\mathbb E(X)=rt,
\qquad
\operatorname{Var}(X)=rt.
$$
:::

### Useful facts: Exponential distribution

Let $X\sim \operatorname{Exp}(\lambda)$, where $X$ shows the time between events in a Poisson process with average number of events $\lambda$.

$$
f_X(x)=
\begin{cases}
0, & x<0,\\
\lambda e^{-\lambda x}, & x\ge 0,
\end{cases}
\qquad
F_X(x)=
\begin{cases}
0, & x<0,\\
1-e^{-\lambda x}, & x\ge 0,
\end{cases}
$$

and

$$
\mathbb E(X)=\frac1\lambda,
\qquad
\operatorname{Var}(X)=\frac1{\lambda^2}.
$$
:::

### Useful facts: Sum of normal distributions

All sums and linear combinations of independent normally distributed random variables are also normally distributed.

Suppose $X_1,\ldots,X_n$ are independent normally distributed random variables, with

$$
X_i\sim N(\mu_i,\sigma_i^2),
\qquad i=1,\ldots,n,
$$

and $a_1,\ldots,a_n,b$ are constants. Then

$$
S=\sum_{i=1}^n a_iX_i+b\sim N(\mu,\sigma^2),
$$

where

$$
\mu=\sum_{i=1}^n a_i\mu_i+b
$$

and

$$
\sigma^2
=
\sum_{i=1}^n a_i^2\sigma_i^2
+
2\sum_{i<j}a_i a_j\operatorname{Cov}(X_i,X_j).
$$

In particular, if $X_1,\ldots,X_n$ are independent, then all covariance terms vanish.
:::

### Useful facts: Chi-square distribution

Let $Z_1,Z_2,\ldots,Z_k$ be independent $N(0,1)$ random variables. If

$$
Q=Z_1^2+Z_2^2+\cdots+Z_k^2=\sum_{i=1}^k Z_i^2,
$$

then the distribution of $Q$ is the $\chi^2$ distribution with $k$ degrees of freedom. This is denoted by

$$
Q\sim \chi^2(k)
\qquad\text{or}\qquad
Q\sim \chi_k^2.
$$

The $\chi_k^2$ distribution is continuous and takes values $x\ge 0$. Its mean and variance are

$$
\mathbb E(Q)=k,
\qquad
\operatorname{Var}(Q)=2k.
$$

If $Q_1,\ldots,Q_m$ are independent and

$$
Q_i\sim \chi^2_{k_i},
\qquad i=1,\ldots,m,
$$

then

$$
Q_1+\cdots+Q_m\sim \chi^2_{k_1+\cdots+k_m}.
$$
:::

### Useful facts: Student's $t$-distribution

Let $Z\sim N(0,1)$ and $Q\sim\chi_k^2$ be independent random variables. Then the random variable

$$
T=\frac{Z}{\sqrt{Q/k}}
$$

has the $t$-distribution with $k$ degrees of freedom, also called Student's $t$ distribution.

We write

$$
T\sim t_k.
$$

Its expectation and variance are

$$
\mathbb E(T)=0,\qquad k>1,
$$

and

$$
\operatorname{Var}(T)=\frac{k}{k-2},
\qquad k>2.
$$

As $k$ increases, the $t$-distribution approaches the standard normal distribution.
:::

### Useful facts: $F$-distribution

Let $Q\sim\chi_k^2$ and $R\sim\chi_p^2$ be independent random variables. Then

$$
F=\frac{Q/k}{R/p}\sim F_{k,p}
$$

has the $F$-distribution with degrees of freedom $k$ and $p$.

By inversion,

$$
\frac1F\sim F_{p,k}.
$$

If $T\sim t_k$, then

$$
T^2\sim F_{1,k}.
$$

The expectation and variance are

$$
\mathbb E(F)=\frac{p}{p-2},
\qquad p>2,
$$

and

$$
\operatorname{Var}(F)
=
\frac{2p^2(k+p-2)}{k(p-2)^2(p-4)},
\qquad p>4.
$$
:::
